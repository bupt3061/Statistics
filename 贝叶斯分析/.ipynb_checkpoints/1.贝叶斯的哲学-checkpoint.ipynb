# 第一章 贝叶斯推断的哲学

**实例1**  
史蒂夫是一个害羞且乐于助人的人，但是他对其他人不太关注。他非常乐见事情处于合理顺序，并且对他的工作非常细心。史蒂夫是农民或图书管理员的概率有多大，已知男性图书管理员的人数只有男性农民的1/20。

假设男性图书管理员符合描述的概率为0.95，男性农民符合描述的概率为0.5，也即P(X|A)=0.95，P(X|~A)=0.5。  
P(A|X)=P(X|A)P(A)/P(X)  
P(A|X)=0.95$\times$(1/21)$\div$(1/21$\times$0.95+20/21$\times$0.5)  
P(A|X)=0.087

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
```

```python
figsize(12.5, 4)                   # 设置图像尺寸
plt.rcParams['savefig.dpi'] = 300  # 图片像素
plt.rcParams['figure.dpi'] = 300   # 分辨率
```

```python
colors = ["#348ABD", "#A60628"]
prior = [1/21., 20/21.]
posterior = [0.087, 1-0.087]
plt.bar([0, .7], prior, alpha=0.70, width=0.25, color=colors[0], label="prior distribution", lw="3", edgecolor=colors[0])
plt.bar([0+0.25, .7+0.25], posterior, alpha=0.70, width=0.25, color=colors[1], label="posterior distribution", lw="3", edgecolor=colors[1])

plt.xticks([0.125, 0.825], ["Librarian", "Farmer"])
plt.title("Prior and posterior probabilities of Steve's occupation")
plt.ylabel("Probability")
plt.legend(loc="upper left")
```

**泊松分布**  
设随机变量Z服从泊松分布，记为:  
Z~Poi(λ)  
当Z取值为k时的概率，用P(Z=k)表示:  
P(Z=k) = λ<sup>k</sup>e<sup>-k</sup>/k!, k=0,1,2...

1.λ可以为任意正数，且随着λ增大，得到大值的概率增大。λ称为泊松分布的强度。

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
figsize(12.5, 4)                   # 设置图像尺寸
plt.rcParams['savefig.dpi'] = 300  # 图片像素
plt.rcParams['figure.dpi'] = 300   # 分辨率
```

```python
a = np.arange(16)
poi = stats.poisson
lambda_ = [1.5, 4.25]
colors = ["#348ABD", "#A60628"]
```

```python
plt.bar(a, poi.pmf(a,lambda_[0]), color=colors[0], label="$\lambda = %.1f$" % lambda_[0], alpha=0.6, edgecolor=colors[0], lw="3")
plt.bar(a, poi.pmf(a,lambda_[1]), color=colors[1], label="$\lambda = %.1f$" % lambda_[1], alpha=0.6, edgecolor=colors[1], lw="3")
plt.xticks(a+0.1, a)
plt.legend()
plt.ylabel("Probability of $k$")
plt.xlabel("$k$")
plt.title("Probability mass function of a Poisson random Variable, differing \$\lambda$ values")
```

2.泊松分布另一个重要性质是，它的期望值等于λ:  
E[Z|λ] = λ

**指数分布**  
设随机变量Z服从指数分布，记为:  
Z~Exp(λ)  
其概率密度函数如下:  
f<sub>z</sub>(z|λ)=λe<sup>-λz</sup>,z≥0  
指数型随机变量的期望值为λ的逆，即:  
E[Z|λ]=1/λ

```python
%matplotlib inline
from IPython.core.pylabtools import figsize
import numpy as np
from matplotlib import pyplot as plt
import scipy.stats as stats
```

```python
figsize(12.5, 4)                   # 设置图像尺寸
plt.rcParams['savefig.dpi'] = 300  # 图片像素
plt.rcParams['figure.dpi'] = 300   # 分辨率
```

```python
a = np.linspace(0, 4, 100)
expon = stats.expon
lambda_ = [0.5, 1]
```

```python
for l, c in zip(lambda_, colors):
    plt.plot(a, expon.pdf(a, scale=1./l), lw=3, color=c, label="$\lambda = %.1f$" % l)
    plt.fill_between(a, expon.pdf(a, scale=1./l), color=c, alpha=.1)

plt.legend()
plt.ylabel("Probability density function at $z$")
plt.xlabel("$z$")
plt.ylim(0, 1.2)
plt.title("Probabilitu density function of an exponential random variable, differing $\lambda$ values")
```
